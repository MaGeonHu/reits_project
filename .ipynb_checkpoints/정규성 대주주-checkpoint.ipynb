```python
import pandas as pd
import warnings
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.stats import shapiro
```

```python
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
stock_all = pd.read_csv('전체기간.csv',index_col = 0)
stock_rise = pd.read_csv('금리인상기간_대주주.csv',index_col = 0)
stock_stop = pd.read_csv('동결기간_대주주.csv',index_col = 0)
```

```python


fig = plt.figure(figsize=(20,15))
for i in range(len(stock_all.columns)):
    ax = fig.add_subplot(3,4,i+1)
    df = stock_all.iloc[:,i]
    stats.probplot(df, dist=stats.norm, plot=ax)
    ax.set_title(stock_all.columns[i])
    shapiro_test = stats.shapiro(df)
    print(f"{stock_all.columns[i]}의 정규성 검정 결과: {shapiro_test[1] :.4f}")
plt.show()
    
```

```python


fig = plt.figure(figsize=(20,15))
for i in range(len(stock_rise.columns)):
    ax = fig.add_subplot(3,4,i+1)
    df = stock_rise.iloc[:,i]
    stats.probplot(df, dist=stats.norm, plot=ax)
    ax.set_title(stock_rise.columns[i])
    shapiro_test = stats.shapiro(df)
    print(f"{stock_rise.columns[i]}의 정규성 검정 결과: {shapiro_test[1] :.4f}")
plt.show()
    
```
