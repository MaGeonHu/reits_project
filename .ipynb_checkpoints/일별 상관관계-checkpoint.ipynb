```python
import warnings
from scipy.stats import pearsonr
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
parents = pd.read_csv('대주주일별수익률.csv',index_col = 0)
child = pd.read_csv('리츠일별수익률.csv',index_col = 0)
```

```python
df = pd.DataFrame()
for i in parents.columns:
    corr_list = []
    p_list = []
    for j in child.columns:
        corr_coef, p_value = pearsonr(parents[i],child[j])
        corr_list.append(corr_coef)
        p_list.append(p_value)
    temp_1 = i + '_corr'
    temp_2 = i + '_p_val'
    df[temp_1] = corr_list
    df[temp_2] = p_list
df.index = child.columns
```

```python
df.to_csv('일별수익률_상관계수.csv')
```
