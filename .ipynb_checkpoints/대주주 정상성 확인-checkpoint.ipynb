```python
import datetime
import numpy as np
import pandas as pd
import yfinance as yf
import warnings
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss
warnings.filterwarnings('ignore')
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
def day_price_ret(ticker_symbol,s_date,e_date):
    stock_symbol = ticker_symbol

    stock_data = yf.download(stock_symbol, start=s_date, end=e_date)
    stock_data = stock_data.set_index(pd.to_datetime(stock_data.index))

    weekly_returns = stock_data['Adj Close'].pct_change().dropna()

    return pd.DataFrame(weekly_returns)
```

```python
def adf_test(timeseries):
    result = adfuller(timeseries)
    statics = result[0]
    p_val = result[1]
    return statics,p_val
```

```python
def kpss_test(timeseries):
    result = kpsstest = kpss(timeseries)
    statics = result[0]
    p_val = result[1]
    return statics , p_val
```

```python
# 변동성 확인
def graph_adf_kpss(s_date,e_date,stock_list,stock_name) :
    adf_result = []
    kpss_result = []
    
    
    fig = plt.figure(figsize = (20,20))
    for i in range(len(stock_list)):
        ax = fig.add_subplot(3,4,i+1)
        df = day_price_ret(stock_list[i],s_date,e_date)['Adj Close']
        adf_result.append(adf_test(df)[1])
        kpss_result.append(kpss_test(df)[1])
        ax.plot(df)
        ax.set_title(stock_name[i])
        ax.axhline(0, 0, 200, color='red', linestyle='--', linewidth=2) 
    plt.show()
    
    for i in range(len(stock_list)):
        if (adf_result[i] < 0.05) and (kpss_result[i] > 0.05) :
            print(f"{stock_name[i]} adf_p_val: {adf_result[i]:.3f}, kpss_result: {kpss_result[i]:.3f} -> 정상성")
        elif (adf_result[i] > 0.05) or (kpss_result[i] < 0.05) :
            print(f"{stock_name[i]} adf_p_val: {adf_result[i]:.3f}, kpss_result: {kpss_result[i]:.3f} -> 비정상성")
        
```

```python
stock_list = ['^KS11','^KQ11','034730.ks' , '023530.ks','267250.ks','006730.kq','086790.ks','395400.ks','330590.ks','357120.ks','404990.ks','417310.ks']
stock_name = ['코스피','코스닥','SK㈜','롯데쇼핑','HD현대','서부티앤디','하나금융지주','SK리츠','롯데리츠','코람코라이프인프라리츠','신한서부티앤디리츠','코람코더원리츠']
```
