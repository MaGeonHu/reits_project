```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import datetime
import numpy as np
import pandas as pd
import yfinance as yf
import warnings
```

```python
def day_price_ret(ticker_symbol,s_date,e_date):
    stock_symbol = ticker_symbol

    stock_data = yf.download(stock_symbol, start=s_date, end=e_date)
    stock_data = stock_data.set_index(pd.to_datetime(stock_data.index))

    weekly_returns = stock_data['Adj Close'].pct_change().dropna()

    return pd.DataFrame(weekly_returns)
```

```python
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
c_inf = ['395400.ks' , '330590.ks','357120.ks','404990.ks','417310.ks']
c_name = ['SK리츠','롯데리츠','코람코라이프인프라리츠','신한서부티앤디리츠','코람코더원리츠']
p_inf = ['034730.ks' , '023530.ks','267250.ks','006730.kq','086790.ks']
p_name = ['SK㈜','롯데쇼핑','HD현대','서부티앤디','하나금융지주']
```

```python
s_date = datetime.datetime(2022,4,1)
e_date = datetime.datetime(2024,4,1)

kos = day_price_ret('^KS11',s_date,e_date)
kod = day_price_ret('^KQ11',s_date,e_date)

fig = plt.figure(figsize = (40,40))
for i in range(len(c_inf)):
    ax = fig.add_subplot(3,5,i+1)
    df = day_price_ret(c_inf[i],s_date,e_date)['Adj Close']
    ax.scatter(df,kos)
    ax.set_title(f"kos - {c_name[i]}")
         
for i in range(len(c_inf)):
    ax = fig.add_subplot(3,5,i+6)
    df = day_price_ret(c_inf[i],s_date,e_date).drop(index =['2022-12-26']).values
    ax.scatter(df,kod)
    ax.set_title(f"kosdaq - {c_name[i]}")


for i in range(len(c_inf)):
    ax = fig.add_subplot(3,5,i+11)
    df = day_price_ret(c_inf[i],s_date,e_date)['Adj Close']
    df1 = day_price_ret(p_inf[i],s_date,e_date)['Adj Close']
    ax.scatter(df,df1)
    ax.set_title(f"{p_name[i]} - {c_name[i]}")
    
plt.show()
```
