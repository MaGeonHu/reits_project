```python
p_range = range(1, 5)  
q_range = range(1, 5)  
p_q_list = []
for i in df.columns:
    results = pd.DataFrame(index=p_range, columns=q_range)
    
    for p in p_range:
        for q in q_range:
            model = arch_model(df.loc[:,i], vol='Garch', p=p, q=q)
            
            result = model.fit(disp='off')
            
            results.loc[p, q] = (result.aic, result.bic)
    
    min_aic = results.stack().idxmin()
    min_bic = results.stack().idxmin()
    p_q_list.append(min_aic)
    
    print(f"{i} - AIC  (p, q):", min_aic)
    print(f"{i} - BIC  (p, q):", min_bic)
```

```python
p_q_df = pd.DataFrame(p_q_list, index = df.columns, columns = ['p','q'])
```

```python
p_q_df.to_csv('p_q_df.csv')
```

```python
var_df = pd.DataFrame()
for i in df.columns:
    p = int(p_q_df.loc[i,'p'])
    q = int(p_q_df.loc[i,'q'])
    result = arch_model(df.loc[:,i], vol='Garch', p=p, q=q).fit(disp='off')
    cond_var = result.conditional_volatility
    residuals = result.resid
    var_df = pd.concat([var_df,cond_var],axis = 1)    
    normality_test = sm.stats.diagnostic.normal_ad(residuals)
    heteroscedasticity_test = sm.stats.diagnostic.het_arch(residuals)
    print(f"{i} Normality test p-value: {normality_test[1]:.4f}")
    print(f"{i} Heteroscedasticity test p-value: {heteroscedasticity_test[1]:.4f}" )
var_df.columns = df.columns
```

```python
var_df.to_csv('.csv')
```
