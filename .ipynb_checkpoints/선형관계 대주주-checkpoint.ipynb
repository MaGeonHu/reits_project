```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import datetime
import numpy as np
import pandas as pd
import yfinance as yf
import warnings
```

```python
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
c_name = ['SK리츠','롯데리츠','코람코라이프인프라리츠','신한서부티앤디리츠','코람코더원리츠']
p_name = ['SK㈜','롯데쇼핑','HD현대','서부티앤디','하나금융지주']
```

```python
stock_all = pd.read_csv('전체기간.csv',index_col=0)

kos = stock_all.loc[:,'코스피']
kod = stock_all.loc[:,'코스닥']

fig = plt.figure(figsize = (30,30))
for i in range(len(c_name)):
    ax = fig.add_subplot(3,5,i+1)
    df = stock_all.loc[:,c_name[i]]
    ax.scatter(df,kos)
    ax.set_title(f"kos - {c_name[i]}")
         
for i in range(len(c_name)):
    ax = fig.add_subplot(3,5,i+6)
    df = stock_all.loc[:,c_name[i]]
    ax.scatter(df,kod)
    ax.set_title(f"kosdaq - {c_name[i]}")


for i in range(len(c_name)):
    ax = fig.add_subplot(3,5,i+11)
    df =  stock_all.loc[:,c_name[i]]
    df1 =  stock_all.loc[:,p_name[i]]
    ax.scatter(df,df1)
    ax.set_title(f"{p_name[i]} - {c_name[i]}")
    
plt.show()
```
