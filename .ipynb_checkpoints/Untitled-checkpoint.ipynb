```python
import pandas as pd
from statsmodels.tsa.vector_ar.vecm import coint_johansen
import warnings
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df2 = pd.read_csv('datasets/리츠일별수익률_real.csv',index_col = 0)
df1 = pd.read_csv('datasets/기업일별수익률_real.csv',index_col = 0)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 대각선에 해당하는 값을 리스트로 생성
values = [0.18, 0.26, 0.05, -0.01, 0.22, 0.16, 0.09]

# 7x7의 0으로 채워진 배열 생성
data = np.zeros((7, 7))

# 대각선에 값 할당
np.fill_diagonal(data, values)

# 히트맵 그리기
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, cmap='coolwarm', fmt=".2f", square=True, xticklabels=df1.columns, yticklabels=df2.columns)
plt.title('Diagonal Heatmap')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
