```python
from scipy.stats import pearsonr
import pandas as pd
from arch import arch_model
import matplotlib.pyplot as plt
import warnings
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
warnings.filterwarnings('ignore')
```

```python
parents = pd.read_csv('./datasets/기업일별수익률_real.csv',index_col = 0)
child = pd.read_csv('./datasets/리츠일별수익률_real.csv',index_col = 0)
```

```python
def adf_test(timeseries):
    result = adfuller(timeseries)
    statics = result[0]
    p_val = result[1]
    return statics,p_val
```

```python
def kpss_test(timeseries):
    result = kpss(timeseries)
    statics = result[0]
    p_val = result[1]
    return statics , p_val
```

```python
def graph_sh(df) :
    fig = plt.figure(figsize = (12,7))
    for i in range(len(df.columns)):
        ax = fig.add_subplot(1,2,i+1)
        ax.plot(df.iloc[:,i])
        ax.set_title(df.columns[i])
        ax.axhline(0, 0, 200, color='red', linestyle='--', linewidth=2)
        a_result = adf_test(df.iloc[:,i])[1]
        k_result = kpss_test(df.iloc[:,i])[1]
        if (a_result < 0.05) and (k_result > 0.05) :
            print(f"{df.columns[i]} adf_p_val: {a_result:.3f}, kpss_result: {k_result:.3f} -> 정상성")
        elif (a_result > 0.05) or (k_result < 0.05) :
            print(f"{df.columns[i]} adf_p_val: {a_result:.3f}, kpss_result: {k_result:.3f} -> 비정상성")
    plt.show()
```

```python
for i in range(len(parents.columns)):
    df = pd.concat([parents.iloc[:,i],child.iloc[:,i]],axis = 1)
    df = df.dropna()
    graph_sh(df)
```

```python
temp = pd.concat([parents.iloc[:,0],child.iloc[:,0]],axis = 1).dropna()
temp1 = pd.concat([parents.iloc[:,1],child.iloc[:,1]],axis = 1).dropna()
temp2 = pd.concat([parents.iloc[:,2],child.iloc[:,2]],axis = 1).dropna()
temp3 = pd.concat([parents.iloc[:,3],child.iloc[:,3]],axis = 1).dropna()
temp4 = pd.concat([parents.iloc[:,4],child.iloc[:,4]],axis = 1).dropna()
temp5 = pd.concat([parents.iloc[:,5],child.iloc[:,5]],axis = 1).diff().dropna()
temp6 = pd.concat([parents.iloc[:,6],child.iloc[:,6]],axis = 1).dropna()
```

```python
df1 = [temp,temp1,temp2,temp3,temp4,temp5,temp6]
```

```python
p_range = range(1, 5)  
q_range = range(1, 5)  
p_q_list = []
for i in df1:
    for j in range(len(i.columns)):
        results = pd.DataFrame(index=p_range, columns=q_range)
        
        for p in p_range:
            for q in q_range:
                model = arch_model(i.iloc[:,j], vol='Garch', p=p, q=q)
                
                result = model.fit(disp='off')
                
                results.loc[p, q] = (result.aic, result.bic)
        
        min_aic = results.stack().idxmin()
        min_bic = results.stack().idxmin()
        p_q_list.append(min_aic)
        
        print(f"{i} - AIC를 최소화하는 (p, q):", min_aic)
        print(f"{i} - BIC를 최소화하는 (p, q):", min_bic)
```

```python
p_q_df = pd.DataFrame(p_q_list , columns = ['p','q'])
coef_list = []
p_val_list = []
```

```python
p_q_df
```

```python
result = arch_model(temp.iloc[:,0], vol='Garch', p=int(p_q_df.iloc[0,0]), q=int(p_q_df.iloc[0,1])).fit(disp='off')
cond_var1 = result.conditional_volatility
result = arch_model(temp.iloc[:,1], vol='Garch', p=int(p_q_df.iloc[1,0]), q=int(p_q_df.iloc[1,1])).fit(disp='off')
cond_var2 = result.conditional_volatility
print(f'{temp.columns[0]},{temp.columns[1]} ')
corr_coef, p_value = pearsonr(cond_var1,cond_var2)
coef_list.append(corr_coef)
p_val_list.append(p_value)
print(corr_coef, p_value)
```
