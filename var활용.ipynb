```python
df = pd.read_csv('전체기간.csv',index_col = 0)
df.head()
```

```python
df1 = df['SK㈜'] * 100
df2 = df['SK리츠'] * 100
```

```python
data = pd.concat([df1,df2], axis = 1)
```

```python
data
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller

# 예시 데이터 생성
np.random.seed(0)


# 차분이 필요한 경우 수행
data_diff = data.diff().dropna()

# VAR 모델 학습 (차분된 데이터를 사용)
model = VAR(data_diff)

# 모델 적합
results = model.fit(maxlags=15, ic='aic')

# # 모델 요약 정보 출력
print(results.summary())
```

```python
print(results.test_causality('SK리츠', ['SK㈜'], kind='f'))
```

```python
data = pd.concat([df['롯데쇼핑'],df['롯데리츠']], axis = 1)
```

```python
data_diff = data.diff().dropna()

# VAR 모델 학습 (차분된 데이터를 사용)
model = VAR(data_diff)

# 모델 적합
results = model.fit(maxlags=15, ic='aic')

# # 모델 요약 정보 출력
print(results.summary())

print(results.test_causality('롯데쇼핑', ['롯데리츠'], kind='f'))
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
test_result = grangercausalitytests(data, 5, verbose=True)
```
