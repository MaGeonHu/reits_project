```python
import pandas as pd
import warnings
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.stats import shapiro
```

```python
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
df1 = pd.read_csv('datasets/기업일별수익률_real.csv', index_col = 0)
df2 = pd.read_csv('datasets/리츠일별수익률_real.csv', index_col = 0)
```

```python
df2
```

```python


fig = plt.figure(figsize=(20,15))
for i in range(7):
    ax = fig.add_subplot(2,4,i+1)
    df = df1.iloc[:,i]
    stats.probplot(df, dist=stats.norm, plot=ax)
    ax.set_title(df1.columns[i])
    shapiro_test = stats.shapiro(df)
    print(f"{df1.columns[i]}의 정규성 검정 결과: {shapiro_test[1] :.4f}")
plt.show()
    
```

```python


fig = plt.figure(figsize=(20,15))
for i in range(len(df1.columns)):
    ax = fig.add_subplot(3,7,i+1)
    df = df1.iloc[:,i]
    df = df.dropna()
    stats.probplot(df, dist=stats.norm, plot=ax)
    ax.set_title(df1.columns[i])
    shapiro_test = stats.shapiro(df)
    print(f"{df1.columns[i]}의 정규성 검정 결과: {shapiro_test[1] :.4f}")
plt.show()
    
```

```python


fig = plt.figure(figsize=(20,15))
for i in range(len(df2.columns)):
    ax = fig.add_subplot(3,7,i+1)
    df = df2.iloc[:,i]
    df= df.dropna()
    stats.probplot(df, dist=stats.norm, plot=ax)
    ax.set_title(df2.columns[i])
    shapiro_test = stats.shapiro(df)
    print(f"{df2.columns[i]}의 정규성 검정 결과: {shapiro_test[1] :.4f}")
plt.show()
    
```
